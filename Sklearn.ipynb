# Линейные модели. Классификация

```python
from matplotlib.colors import ListedColormap
from sklearn import model_selection, datasets, linear_model, metrics

import numpy as np

%pylab inline
```

## Генерация данных

```python
blobs = datasets.make_blobs(centers=2, cluster_std=5.5, random_state=1)
```

```python
colors = ListedColormap(['red', 'black'])

plot_2d_datset(blobs, colors)
```

```python
train_data, test_date, train_lables, test_labels = model_selection.train_test_split(
    blobs[0], blobs[1], test_size = 0.3, random_state=1) 
```

## Линейная классификация

### RidgeClassifier

```python
# создание объекта - классификатора
ridge_classifier = linear_model.RidgeClassifier(random_state=1)
```

```python
# обучение классификатора
ridge_classifier.fit(train_data, train_lables)
```

```python
# применение обученного классификатора
ridge_predictions = ridge_classifier.predict(test_date)
```

```python
test_labels
```

```python
ridge_predictions
```

```python
# оценка качсества классификации
metrics.accuracy_score(test_labels, ridge_predictions)
```

```python
ridge_classifier.coef_
```

```python
ridge_classifier.intercept_
```

```python
# -0.0854443x + -0.07273219y + -0.31250723 = 0
```

```python
k = -0.0854443 / 0.07273219
b = -0.31250723 / 0.07273219
```

```python
t = np.linspace(-20, 10, 100)
plot_2d_datset(blobs, colors)
plt.plot(t, k*t+b)
```

### Логистическая регрессия

```python
# создание объекта - классификатора
logistic_classifier = linear_model.LogisticRegression(random_state=1)
```

```python
# обучение объекта - классификатора
logistic_classifier.fit(train_data, train_lables)
```

```python
# применение обученного классификатора
logistic_predicted = logistic_classifier.predict(test_date)
logistic_predicted_proba = logistic_classifier.predict_proba(test_date)
```

```python
logistic_predicted
```

```python
test_labels
```

```python
# оценка качсества классификации
metrics.accuracy_score(logistic_predicted, test_labels)
```

## Оценка качества по cross-validation

### cross_val_score

```python
ridge_scoring = model_selection.cross_val_score(ridge_classifier, blobs[0], blobs[1], scoring='accuracy', cv=10)
```

```python
lr_scoring = model_selection.cross_val_score(logistic_classifier, blobs[0], blobs[1], scoring='accuracy', cv=10)
```

```python
ridge_scoring.mean(), ridge_scoring.max(), ridge_scoring.min()
```

```python
lr_scoring.mean(), lr_scoring.max(), lr_scoring.min()
```

```python
lr_scoring
```

```python
ridge_scoring
```

### cross_val_score с заданными scorer и cv_strategy

```python
scorer = metrics.make_scorer(metrics.accuracy_score)
```

```python
cv_strategy = model_selection.StratifiedShuffleSplit(n_splits=20, test_size=0.3, random_state=2)
```

```python
ridge_scoring = model_selection.cross_val_score(ridge_classifier, blobs[0], blobs[1], scoring=scorer, cv=cv_strategy)
```

```python
lr_scoring = model_selection.cross_val_score(logistic_classifier, blobs[0], blobs[1], scoring=scorer, cv=cv_strategy)
```

```python
ridge_scoring.mean(), ridge_scoring.max(), ridge_scoring.min()
```

```python
lr_scoring.mean(), lr_scoring.max(), lr_scoring.min()
```

# Метрики качества

```python
from matplotlib.colors import ListedColormap
from sklearn import model_selection, datasets, linear_model, metrics

%pylab inline
```

```python
clf_data, clf_target = datasets.make_classification(n_features=2, 
                                                    n_informative=2, 
                                                    n_classes=2, 
                                                    n_redundant=0, 
                                                   n_clusters_per_class=1,
                                                   random_state=7)
```

```python
reg_data, reg_target = datasets.make_regression(n_features=2, n_informative=1, n_targets=1, noise=5., random_state=7)
```

```python
colors = ListedColormap(['red', 'black'])
pylab.scatter(list(map(lambda x: x[0], clf_data)), list(map(lambda x: x[1], clf_data)), c=clf_target, cmap=colors)
```

```python
pylab.scatter(list(map(lambda x: x[1], reg_data)), reg_target, color='r')
pylab.scatter(list(map(lambda x: x[0], reg_data)), reg_target, color='b')
```

```python
clf_train_data, clf_test_date, clf_train_lables, clf_test_labels = model_selection.train_test_split(
    clf_data, clf_target, test_size = 0.3, random_state=1)
```

```python
reg_train_data, reg_test_date, reg_train_lables, reg_test_labels = model_selection.train_test_split(
    reg_data, reg_target, test_size = 0.3, random_state=1)
```

## Метрики качества в задачах классификации

### Обучение модели классификации

```python
classifier = linear_model.SGDClassifier(loss='log', random_state=1)
```

```python
classifier.fit(clf_train_data, clf_train_lables)
```

```python
predictions = classifier.predict(clf_test_date)
```

```python
probability_predictions = classifier.predict_proba(clf_test_date)
```

```python
clf_test_labels
```

```python
predictions
```

```python
probability_predictions
```

### Метрики в задачах классификации

```python
## accuracy
metrics.accuracy_score(clf_test_labels, predictions)
```

```python
## confusion matrix - матрица ошибок
metrics.confusion_matrix(clf_test_labels, predictions)
```

```python
## precision
metrics.precision_score(clf_test_labels, predictions, pos_label=0)
```

```python
metrics.precision_score(clf_test_labels, predictions)
```

```python
## f мера
metrics.f1_score(clf_test_labels, predictions)
```
